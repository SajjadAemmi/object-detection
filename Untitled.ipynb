```python
import cv2
import numpy as np
import os
import tensorflow as tf
import glob
from matplotlib import pyplot as plt
from object_detection.utils import ops as utils_ops
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as vis_util
```

```python
detection_model = tf.saved_model.load(str('inference_graph_tf1/uint8_1/saved_model'))
model_fn = detection_model.signatures['serving_default']
```

```python
frame = cv2.imread('/media/sajjad/08A41DFAA41DEB48/Dataset/farazist/16- نوشابه 300 کوکاکولا/2019-08-29_15-27-38.jpg')
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

input_tensor = tf.convert_to_tensor(frame)
# The model expects a batch of images, so add an axis with `tf.newaxis`.
input_tensor = input_tensor[tf.newaxis,...]

# Run inference

output_dict = model_fn(input_tensor)

# All outputs are batches tensors.
# Convert to numpy arrays, and take index [0] to remove the batch dimension.
# We're only interested in the first num_detections.
num_detections = int(output_dict.pop('num_detections'))
output_dict = {key:value[0, :num_detections].numpy() 
                for key,value in output_dict.items()}
```

```python
index = np.argmax(output_dict['detection_scores'])
box = output_dict['detection_boxes'][index]

r, c, _ = frame.shape

xmin = int(box[0] * r)
ymin = int(box[1] * c)
xmax = int(box[2] * r)
ymax = int(box[3] * c)

image = frame[xmin:xmax, ymin:ymax]

%matplotlib inline
plt.imshow(image)
plt.show()
```

```python
source_path = '/media/sajjad/08A41DFAA41DEB48/Dataset/farazist_pre'
destination_path = '/media/sajjad/08A41DFAA41DEB48/Dataset/farazist_pre'

for id, d in enumerate(os.listdir(source_path)):
    if os.path.isdir(os.path.join(source_path, d)):
        for f in glob.glob(os.path.join(source_path, d, "*.jpg")):
            
            frame = cv2.imread(os.path.join(source_path, d, f))
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            input_tensor = tf.convert_to_tensor(frame)
            # The model expects a batch of images, so add an axis with `tf.newaxis`.
            input_tensor = input_tensor[tf.newaxis,...]

            # Run inference

            output_dict = model_fn(input_tensor)

            # All outputs are batches tensors.
            # Convert to numpy arrays, and take index [0] to remove the batch dimension.
            # We're only interested in the first num_detections.
            num_detections = int(output_dict.pop('num_detections'))
            output_dict = {key:value[0, :num_detections].numpy() 
                            for key,value in output_dict.items()}
            
            index = np.argmax(output_dict['detection_scores'])
            box = output_dict['detection_boxes'][index]

            r, c, _ = frame.shape

            xmin = int(box[0] * r)
            ymin = int(box[1] * c)
            xmax = int(box[2] * r)
            ymax = int(box[3] * c)

            image = frame[xmin:xmax, ymin:ymax]
            
            if not os.path.exists(os.path.join(destination_path, d)):
                os.makedirs(os.path.join(destination_path, d))
            
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            cv2.imwrite(os.path.join(destination_path, d, f), image)
```
